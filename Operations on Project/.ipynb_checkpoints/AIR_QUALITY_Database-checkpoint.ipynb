## ***---> CLEANING DATA***

## **1. Handling Missing Data**

```python
air_quality_data.replace("?",np.NaN, inplace=True)
```

Find Missing Data

```python
air_quality_data.isnull().sum()
for column in air_quality_data.columns:
    print(column)
    print(air_quality_data[column].isnull().sum())
```

Handle Missing Data

```python
#filling the empty message rows with categorizing it as Good or Bad quality
for idx, entry in air_quality_data['Data Value'].items():
    if entry > 25:
        message = 'Bad Quality'
        air_quality_data.loc[idx, "Message"] = message
    else:
        message = 'Good Quality'
        air_quality_data.loc[idx, "Message"] = message
```

## **---> Performing SQL Operations**

### 1. Extract all the name of the gases present in the Geo ID - '102'

```python
%sql SELECT DISTINCT(name_gas) FROM AIR_QUALITY WHERE geo_join_id = 102
```

### 2. Group the various time periods to figure out how many records have been taken in each of the time periods.

```python
%sql SELECT time_period, COUNT(unique_id) FROM AIR_QUALITY GROUP BY time_period
```

### 3. Filter and order the records with geotype name as 'UHF42' and measure info as 'ppb' based on their data value.

```python
%%sql SELECT unique_id, geo_type_name, measure_info, data_value
FROM AIR_QUALITY
WHERE geo_type_name = 'UHF42' AND measure_info = 'ppb'
ORDER BY data_value
```

### 4. Select the records in which geo place names data value percentage is above 60 %

```python
%%sql SELECT unique_id, geo_place_name, data_value_percentage
FROM AIR_QUALITY
WHERE data_value_percentage > 60
```

### 5. Create a procedure to find the average data value of a particular gas

```python
%%sql CREATE OR REPLACE FUNCTION avg_data_values(gas_name varchar(100))
	RETURNS numeric(8,5) AS $$
	DECLARE
		Avg_value numeric(8,5);
	BEGIN
		SELECT AVG(data_value) INTO Avg_value
		FROM AIR_QUALITY
		WHERE name_gas = gas_name;
		RETURN Avg_value;
	END;
$$ language plpgsql;

SELECT * FROM avg_data_values('Nitrogen dioxide (NO2)') AS average;
```

## **Visualization of Data**

### 1. Correlation between features

```python
AIR_QUALITY_DATAv = pd.read_csv('/Users/rohanniranjankalpavruksha/Desktop/ROHAN/Pace University/SEM 1/Scalable Database/Project-MidTerm/Cleaned_Air_Quality_Dataset.csv')
```

```python
AIR_QUALITY_DATAv.head()
```

```python
AIR_QUALITY_DATAv['Start_Date']= pd.to_datetime(AIR_QUALITY_DATAv['Start_Date'])
```

```python
# Import seaborn
import seaborn as sns
corr = AIR_QUALITY_DATAv[[ 'Unique ID', 'Indicator ID', 'Geo Join ID', 'Start_Date', 'Data Value', 'Data Value Percentage']].corr()
ax = sns.heatmap(
    corr,
    vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(20, 220, n=200),
    square=True
)
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=45,
    horizontalalignment='right'
);
```

```python
df=pd.DataFrame(AIR_QUALITY_DATAv)
df['Name'] =df['Name'].astype('category').cat.codes         #Handling Categorical Features
df['Measure'] =df['Measure'].astype('category').cat.codes
df['Measure Info'] =df['Measure Info'].astype('category').cat.codes
df['Geo Type Name'] =df['Geo Type Name'].astype('category').cat.codes
df['Geo Place Name'] =df['Geo Place Name'].astype('category').cat.codes
df['Time Period'] =df['Time Period'].astype('category').cat.codes
df['Message'] =df['Message'].astype('category').cat.codes
```
