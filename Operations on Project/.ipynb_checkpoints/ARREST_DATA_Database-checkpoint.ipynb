## ***---> Importing Libraries***

```python
import numpy as np                #library used to perform mathematical operations on arrays
import pandas as pd               #library used for working with data sets
import matplotlib.pyplot as plt   #library for interactive visualizations
```

## ***---> Extracting the Dataset File***

```python
arrest_data = pd.read_csv('/Users/rohanniranjankalpavruksha/Desktop/ROHAN/Pace University/SEM 1/Scalable Database/Project-MidTerm/Cleaned_Arrest_Data_Dataset.csv')("/content/")    #raw input dataset
```

## ***---> CLEANING DATA***

## **1. Handling Missing Data**

```python
arrest_data.replace("?",np.NaN, inplace=True)
```

### Find Missing Data

```python
arrest_data.isnull().sum()
for column in arrest_data.columns:
    print(column)
    print(arrest_data[column].isnull().sum())
```

### Handle Missing Data

```python
#replacing null values in PD_CD with most frequent PD_CD
most_frequent_PD_CD = arrest_data["PD_CD"].value_counts().idxmax()
arrest_data["PD_CD"].replace(np.nan, most_frequent_PD_CD, inplace=True)
#replacing null values in KY_CD with most frequent KY_CD
most_frequent_KY_CD = arrest_data["KY_CD"].value_counts().idxmax()
arrest_data["KY_CD"].replace(np.nan, most_frequent_KY_CD, inplace=True)
#replacing null values in LAW_CAT_CD with most frequent LAW_CAT_CD
most_frequent_LAW_CAT_CD = arrest_data["LAW_CAT_CD"].value_counts().idxmax()
arrest_data["LAW_CAT_CD"].replace(np.nan, most_frequent_LAW_CAT_CD, inplace=True)
#replacing null values in Zip Codes with most frequent Zip Codes
most_frequent_zip_codes = arrest_data["Zip Codes"].value_counts().idxmax()
arrest_data["Zip Codes"].replace(np.nan, most_frequent_zip_codes, inplace=True)

#replacing null values in Community Districts with mean
average_community_districts = arrest_data["Community Districts"].astype("float").mean()
arrest_data["Community Districts"].replace(np.nan, average_community_districts, inplace= True)
#replacing null values in Borough Boundaries with mean
average_borough_boundaries = arrest_data["Borough Boundaries"].astype("float").mean()
arrest_data["Borough Boundaries"].replace(np.nan, average_borough_boundaries, inplace= True)
#replacing null values in City Council Districts with mean
average_city_council_districts = arrest_data["City Council Districts"].astype("float").mean()
arrest_data["City Council Districts"].replace(np.nan, average_city_council_districts, inplace= True)
#replacing null values in Police Precincts with mean
average_police_precincts = arrest_data["Police Precincts"].astype("float").mean()
arrest_data["Police Precincts"].replace(np.nan, average_police_precincts, inplace= True)
```

## **2. Correcting Data Format**

```python
arrest_data.dtypes    #listing data types for each column
```

```python
#converting the data types to correct format
arrest_data[['Community Districts', 'Borough Boundaries', 'City Council Districts', 'Police Precincts', 'Zip Codes']] = arrest_data[['Community Districts', 'Borough Boundaries', 'City Council Districts', 'Police Precincts', 'Zip Codes']].astype('int')
arrest_data.dtypes #checking it again
```

## **3. Normalizing Data**

```python
#normalizing Latitude
arrest_data["X_COORD_CD"] = arrest_data["X_COORD_CD"]/arrest_data["X_COORD_CD"].max()
arrest_data["Y_COORD_CD"] = arrest_data["Y_COORD_CD"]/arrest_data["Y_COORD_CD"].max()
```

## **---> Performing SQL Operations**

### 1. Figure out the count of each sex/gender arrested

```python
%%sql SELECT perp_sex, COUNT(perp_sex)
FROM ARREST_DATA
GROUP BY perp_sex
```

### 2. Figure out for what reason are people of age 25-44 are majorly accused of

```python
%%sql SELECT DISTINCT(ofns_desc)
FROM ARREST_DATA
WHERE age_group = '25-44'
```

### 3. What are the top 5 reasons being arrested and their count

```python
%%sql SELECT ofns_desc, COUNT(arrest_key)
FROM ARREST_DATA
GROUP BY ofns_desc
ORDER BY COUNT(arrest_key) DESC
LIMIT 5
```

### 4. Arrange the Borough Boundaries based on the Crime Rates

```python
%%sql SELECT borough_boundaries, COUNT(arrest_key)
FROM ARREST_DATA
GROUP BY borough_boundaries
ORDER BY COUNT(arrest_key)
```

### 5. Create a Procedure to return the area in which the offence has happened based on arrest key

```python
%%sql CREATE OR REPLACE FUNCTION address_of_crime(arrestKey numeric(15))
  RETURNS TABLE (Lat numeric(15,10), Long numeric(15,10), Zip numeric(8,2))
  LANGUAGE plpgsql AS
$func$
BEGIN
   RETURN QUERY
   SELECT latitude, longitude, zip_codes
   FROM ARREST_DATA
   WHERE arrest_key = arrestKey;
END
$func$;

SELECT * FROM address_of_crime(261249590)
```
